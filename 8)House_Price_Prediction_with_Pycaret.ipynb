```python
#Import necessary modules
import pandas as pd
from pycaret.regression import *
```

```python
#Load data
Caf_house_df = pd.read_csv('california_housing_train.csv')
Caf_house_df.head(10)
```

```python
TM = setup(Caf_house_df, target='median_house_value')
```

```python
best = compare_models()
```

```python
#Explore the best model parameter
print(best)
```

```python
finalize_model(best)
```

```python
evaluate_model(best)
```

```python
predict_model(best)
```

```python
calif_test_df = pd.read_csv('california_housing_train.csv')

predict_model(best, calif_test_df)
```
